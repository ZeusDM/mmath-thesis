```python
from sage.combinat.cartesian_product import CartesianProduct_iters

S = ['a', 'b', 'c', 'd']

# all set compositions
comps = [
    [set(block) for block in order]
    for P in SetPartitions(S)
    for order in Permutations(P)
]

# build all (C, l) pairs
pairs = []
for C in comps:
    choices = []
    for block in C:
        if len(block) == 1:
            choices.append([1])
        else:
            choices.append([1, -1])
    for l in CartesianProduct_iters(*choices):
        pairs.append((C, list(l)))


def chi(C, l, x, y):
    i = next(idx for idx, block in enumerate(C) if x in block)
    j = next(idx for idx, block in enumerate(C) if y in block)
    if i > j:
        return True
    if i == j and l[i] == -1:
        return True
    return False


# keep only those with odd parity condition
odd_pairs = []
for C, l in pairs:
    val = (
        chi(C, l, 'a', 'c')
        + chi(C, l, 'a', 'd')
        + chi(C, l, 'b', 'c')
        + chi(C, l, 'b', 'd')
    ) % 2
    if val == 1:
        odd_pairs.append((C, l))


def composition_to_string(C, l):
    """Return a LaTeX-ready comparison chain for elements of S = [a,b,c,d]."""
    blocks = []
    for block, sign in zip(C, l):
        block = sorted(list(block))
        if len(block) == 1:
            blocks.append(block[0])
        else:
            symbol = f"\\overset{{{'+' if sign == 1 else '-'}}}{{=}}"
            part = f" {symbol} ".join(block)
            blocks.append(part)
    return " > ".join(blocks)

# print results in requested format
for C, l in odd_pairs:
    print(composition_to_string(C, l), "\\\\")
```

```python
def chi3(C, l, x, y, z):
    """Ternary chi(x, y, z) = chi(x, y) + chi(y, z) - chi(x, z)."""
    return chi(C, l, x, y) + chi(C, l, y, z) - chi(C, l, x, z)


def in_different_blocks(C, x, y):
    """Check if x and y are in different parts of C."""
    i = next(idx for idx, block in enumerate(C) if x in block)
    j = next(idx for idx, block in enumerate(C) if y in block)
    return i != j

# find all pairs satisfying the new constraints
valid_pairs = []
for C, l in pairs:
    if not (in_different_blocks(C, 'a', 'b') and in_different_blocks(C, 'c', 'd')):
        continue
    cabc = chi3(C, l, 'a', 'b', 'c')
    cbcd = chi3(C, l, 'b', 'c', 'd')
    cadc = chi3(C, l, 'a', 'd', 'c')
    if cabc == cbcd == cadc:
        valid_pairs.append((C, l))


def composition_to_string(C, l):
    """Return a LaTeX-ready comparison chain for elements of S = [a,b,c,d]."""
    blocks = []
    for block, sign in zip(C, l):
        block = sorted(list(block))
        if len(block) == 1:
            blocks.append(block[0])
        else:
            symbol = f"\\overset{{{'+' if sign == 1 else '-'}}}{{=}}"
            part = f" {symbol} ".join(block)
            blocks.append(part)
    return " > ".join(blocks)


# print results
for C, l in valid_pairs:
    print(composition_to_string(C, l), "\\\\")
```
